рдЕрдм рд╡рд╣реА рдХрд╛рдо рд╣рдо рд╕рд╛рдзрд╛рд░рдг Python рд▓реЙрдЬрд┐рдХ рд╕реЗ рдХрд░реЗрдВрдЧреЗред
(рдпрд╣рд╛рдБ рди AI рд╣реИ, рди рдорд╢реАрди рд▓рд░реНрдирд┐рдВрдЧ)

рд╕рд░рд▓ рд╡рд┐рдЪрд╛рд░:

тАЬрдФрд╕рдд рдмрджрд▓рд╛рд╡ рдирд┐рдХрд╛рд▓реЛ
рдФрд░ рдЙрд╕реА рдХреЗ рдЖрдзрд╛рд░ рдкрд░ рднрд╡рд┐рд╖реНрдп рдХрд╛ рдЕрдиреБрдорд╛рди рд▓рдЧрд╛рдУредтАЭ

```python
# Past data
time = [1, 2, 3, 4]
money = [50, 100, 150, 200]

# Har ghante kitna paisa badh raha hai
increase_per_hour = []

for i in range(1, len(time)):
    change = money[i] - money[i-1]
    increase_per_hour.append(change)

average_increase = sum(increase_per_hour) / len(increase_per_hour)

# Future prediction
future_time = 5
predicted_money = average_increase * future_time

print(predicted_money)
```

рд╕рд╛рдзрд╛рд░рдг рдФрд░ рдмрдЪреНрдЪреЛрдВ рдЬреИрд╕реА рд╣рд┐рдВрджреА рдореЗрдВ (рд▓рд╛рдЗрди-рдмрд╛рдп-рд▓рд╛рдЗрди):
*   time тЖТ рдЗрдирдкреБрдЯ рдбреЗрдЯрд╛
*   money тЖТ рдЖрдЙрдЯрдкреБрдЯ рдбреЗрдЯрд╛
*   Loop тЖТ рд╣рд░ рд╕реНрдЯреЗрдк рдкрд░ рдлрд░реНрдХ рдирд┐рдХрд╛рд▓рдирд╛
*   Average тЖТ рдПрдХ рдЬреИрд╕рд╛ рдирд┐рдпрдо рдмрдирд╛рдирд╛
*   Multiply тЖТ рднрд╡рд┐рд╖реНрдп рдХрд╛ рдЕрдВрджрд╛рдЬрд╝рд╛ рд▓рдЧрд╛рдирд╛
